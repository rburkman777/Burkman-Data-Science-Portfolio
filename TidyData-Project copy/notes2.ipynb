We are going to explore data on olmpyic medalists in the 2008 olympics and examine how best to craft code to clean it up and make it tidy. First, let's import our data. 

```python
import pandas as pd

# First, we use the read_csv function to load the data into our DataFrame

df = pd.read_csv('olympics_08_medalists.csv') 
```

Next, we will melt the data frame to better organize it. Currently, we have many columns and the data needs to be simplified into having one column for all of the these variables.  

```python
#this allows us to grab the columns
value_vars = [col for col in df.columns if col != 'medalist_name']

#next, we "melt" the data so that the many columns for each sport become one column. This makes the data far more organized. We also create a column for medal type here.
df_melted = pd.melt(
    df,
    id_vars=['medalist_name'],
    value_vars=value_vars,
    var_name='Sport',
    value_name='Medal'
)
```
